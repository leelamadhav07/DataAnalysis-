# Seaborn - Statistical Data Visualization


### What is Seaborn ?

Seaborn is a python library, that builds on the top of matplotlib library, that makes the statistical data for easy and beautiful.

### Why we need it ? 

1) For creating plots like violin plots, box plots, plain plots.

2) Automatically handles data from pandas data frames.

### Where we use it ?

1) Data Analysis and Re-Booting.

2) Exploring relationships between variables

3) Machine learning -- For visualizing data sets and results.

```python
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
```

```python
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
###  Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
```

```python
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
iris_data = load_iris(as_frame=True)
iris = iris_data.frame
iris['species'] = iris_data.target_names[iris['target']]
sns.pairplot(iris, hue="species")
plt.show()
```

```python
corr = iris.drop(columns=['species']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset (Seaborn has it built-in)
titanic = sns.load_dataset('titanic')

# Display first few rows
print(titanic.head())

```

```python
#  Compare Passenger Survival by Class and Gender
plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x='class', hue='sex')
plt.title('Passenger Distribution by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=titanic, x='class', y='survived', hue='sex')
plt.title('Passenger Survival Rate by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#  Distribution of Passenger Ages
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x='age', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Fare Distribution by Class
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='class', y='fare', palette='Set2')
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
#  Relationship Between Age and Fare
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x='age', y='fare', hue='class', style='sex')
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Class')
plt.show()
```

```python
#  Correlation Heatmap
plt.figure(figsize=(8,5))
corr = titanic[['survived', 'pclass', 'age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Titanic Dataset)')
plt.show()
```
